# PatchCore 
- "사전에 학습된 CNN(Resnet)으로 정상 이미지들의 패치 (작은영역)feature를 뽑아서 메모리 뱅크에 저장해두고, 테스트 이미지의 패치 feature가 그 메모리 뱅크의 feature들과 얼마나 멀리 떨어져 있는지(거리)를 계산해서 비정상 여부를 판단

- 이미지를 직접 복원하는 게 아니라 "이미 학습된 좋은 feature 공간에서, 정상 패턴들과 얼마나 다른가"를 확인
- 학습을 따로 하지 않음. Pretrained CNN을 두고, "정상 이미지들의 feature를 모아서 저장"
